```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
```

```python
url = 'https://raw.githubusercontent.com/MainakRepositor/Datasets-/master/Suicide%20data.csv'
df = pd.read_csv(url , error_bad_lines=False)
```

```python
df.isna().sum()
```

```python
df = df.drop(columns = ['country-year' , 'HDI for year'])
```

```python
df.isna().sum()
```

```python
df.describe().plot(kind = "area",fontsize=15, figsize = (17,7), table = True,colormap="rainbow")
plt.xlabel('Statistics',)
plt.ylabel('Value')
plt.title("General Statistics")
plt.show()
```

```python
plt.figure(figsize = (16,7))
cor = sns.heatmap(df.corr(), annot = True,cmap='hot_r')
plt.title("Spearman correlation matrix for parametes\n",size=20)
```

```python
plt.figure(figsize = (15,25))
sns.countplot(y = 'country' , data = df , alpha = 0.9)
plt.title('DATA BY COUNTRY' , size= 20)
plt.show()
```

```python
plt.figure(figsize = (15,8))
sns.countplot(x = 'sex' , data = df )
plt.show()
```

```python
plt.figure(figsize=(16,7))
bar_age = sns.barplot(x = 'sex', y = 'suicides_no', hue = 'age',data = df)
```

```python
plt.figure(figsize=(16,7))
bar_age = sns.barplot(x = 'sex', y = 'suicides_no', hue = 'generation',data = df)
```

```python
cat_acc_yr = sns.catplot(x = 'sex' , y = 'suicides_no' , hue = 'age' , col = 'year' , data=df , kind='bar' , col_wrap=3)
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['generation'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Generations Count',size=18)
ax[0].set_ylabel('Count')
sns.countplot(x='generation',data=df,ax=ax[1])
ax[1].set_title('Generations Count',size=18)
plt.show()
```
